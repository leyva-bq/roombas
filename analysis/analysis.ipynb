# Analisis de tiempo con roombas

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
times = pd.read_csv('times.csv')
```

```python
times_means = times.groupby(['m', 'n', 'numRoombas']).mean()
times_means.reset_index(inplace=True)
times_means
```

```python
times_means['m*n'] = times_means['m'].astype(str) + ' x ' + times_means['n'].astype(str)
times_means
```

## plt construction

```python
sizes = times_means['m*n'].unique()
values = times_means[['numRoombas', 'tiempo']]

x = np.arange(len(sizes))
width = 0.25
multiplier = 0

fig, ax = plt.subplots()

for i in values['numRoombas'].unique():
    ax.bar(x + multiplier - width,
           values[values['numRoombas'] == i]['tiempo'],
           width,
           label=f'{int(i)} Roombas')
    multiplier += width
    
ax.set_ylabel('Tiempo (s)')
ax.set_title('Tiempo de ejecución en función del tamaño del tablero')
ax.set_xticks(x)
ax.set_xticklabels(sizes)
ax.legend()

fig.tight_layout()
plt.show()
```

```python
sizes = times_means['m*n'].unique()
values = times_means[['numRoombas', 'movimientos']]

x = np.arange(len(sizes))
width = 0.25
multiplier = 0

fig, ax = plt.subplots()

for i in values['numRoombas'].unique():
    ax.bar(x + multiplier - width,
           values[values['numRoombas'] == i]['movimientos'],
           width,
           label=f'{int(i)} Roombas')
    multiplier += width
    
ax.set_ylabel('Tiempo (s)')
ax.set_title('Movimientos realizados en función del tamaño del tablero')
ax.set_xticks(x)
ax.set_xticklabels(sizes)
ax.legend()

fig.tight_layout()
plt.show()
```
